```python
#Task-1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?
Ans:In NLP, "Love" and "love" are considered different words because computers treat uppercase and lowercase letters separately.
    This is called case sensitivity.
    To avoid duplication and improve text analysis, NLP preprocessing usually converts all words to lowercase.

2. What happens if stopwords are not removed?
Ans:If stopwords are not removed, the dataset will contain many common words like 'is', 'the', and 'and' that do not add much meaning. 
    This increases the size of the data and processing time. 
    It may also reduce the efficiency of NLP models.

3. Provide two real-world scenarios where removing stopwords can be harmful.
Ans:Removing stopwords can sometimes change the actual meaning of a sentence.
    For example, in sentiment analysis removing the word 'not' from 'not good' changes the sentiment. 
    It can also affect question-answer systems, where words like 'what' or 'who'are important.

4. Explain the difference between stemming and lemmatization.
Ans:Stemming-reduces a word to its root form by removing suffixes using simple rules, and the result may not always be a real word. 
    Lemmatization-converts a word to its correct dictionary base form using linguistic analysis. 
    Lemmatization is usually more accurate but slower than stemming.
    
```

```python
#Task-2 Build Advanced Preprocessing Function

import re
import string

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words (≤2) except "no", "not"
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence

print("   [Tokens]         Cleaned_text")
print(preprocess_text("I have 2 dogs"))
print(preprocess_text("soooo goooood!!!"))
print(preprocess_text("WOW!!! This is GREAT!!!"))
print(preprocess_text("Visit http://example.com now"))
print(preprocess_text("Nooooo this is baaad!!!"))
```

```python
#Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for sentence in test_sentences:
    tokens, clean_text = preprocess_text(sentence)
    all_tokens.extend(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_text)
    print("-" * 50)
```

```python
#Task 4: Token Analytics

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total if total > 0 else 0

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Length:", round(avg_len, 2))
    print("-" * 50)


# Analysis
most_noisy_sentence = min(test_sentences, key=lambda s: len(preprocess_text(s)[0]))
most_meaningful_sentence = max(test_sentences, key=lambda s: len(preprocess_text(s)[0]))

print("Most noisy sentence:", most_noisy_sentence)
print("Most meaningful sentence:", most_meaningful_sentence)
```

```python
#Task 5: Frequency Analysis

from collections import Counter

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

```python
#Task 6: Build Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Test
result = full_pipeline(test_sentences)
print(result)
```
